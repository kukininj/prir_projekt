```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
titles = {'wyniki_1_average_10.txt': 'rednia z 10 uruchomie - 1 prba', 'wyniki_2_average_10.txt': 'rednia z 10 uruchomie - 2 prba', 'wyniki_average_20.txt': 'rednia z 20 uruchomie'}
for file in titles.keys():
    df = pd.read_csv(file)
    fig, ax = plt.subplots()
    ax.set_title(titles[file])
    ax.set_xlabel('Ilo wtkw')
    ax.set_ylabel('Przyspieszenie')
    ax.set_yticks(np.arange(25))
    ax.set_xticks(df['ilosc watkow'])
    ax.grid(True)
    ax.plot(df['ilosc watkow'], df['czas'][0] / df['czas'], '-o')
    fig.savefig(f'{titles[file]}.png')
```
